```python
import torch
import torch.nn as nn

from src.build_optimizer import build_optimizer
from src.seed import set_seed
```

```python
class Model(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(in_features=2, out_features=1)

    def forward(self, x):
        return self.linear(x)

def test_optimization_loop(model, optimizer, steps = 100):
    set_seed(42)

    X = torch.randn(100, 2)
    true_w = torch.tensor([[3.0, -2.0]])
    true_b = torch.tensor([1.0])
    Y_true = X @ true_w.T + true_b
    
    print("Before train")
    print("Weights:", model.linear.weight.data.numpy())
    print("Bias:", model.linear.bias.data.numpy())
    print()

    criterion = nn.MSELoss()

    for step in range(1, steps + 1):
        # Forward pass
        Y_pred = model(X)
        loss = criterion(Y_pred, Y_true)
        
        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        
        # optimizer step
        optimizer.step()

        if step % 10 == 0:
            print(f"Шаг {step:02d} | Loss: {loss.item():.6f}")

    print("After train")
    print("Target: Weights [[3.0, -2.0]], Bias [1.0]")
    print("Fact: Weights", model.linear.weight.data.numpy())
    print("Fact: Bias", model.linear.bias.data.numpy())
```

```python
model = Model()
optimizer = build_optimizer(
    model=model,
    muon_lr=0.1,
    adamw_lr=0.1
)
test_optimization_loop(model, optimizer)
```
